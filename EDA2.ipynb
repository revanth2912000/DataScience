# Encoding

```python
categories=data.select_dtypes(include='object').columns
categories
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
cols=[col for col in categories if data[col].nunique()<5] 
cols
```

```python
for i in cols:
    encoded=encoder.fit_transform(data[[i]])
    encodedNames=[f'{i}_{j}' for j in encoder.categories_[0] ]
    encodedData=pd.DataFrame(encoded,columns=encodedNames)
    data.drop(columns=[i],inplace=True)
    data=pd.concat([data,encodedData],axis=1)
```

```python
data.isna().sum().sum()
```

```python
data.dropna(inplace=True)
data.isna().sum().sum()
```
