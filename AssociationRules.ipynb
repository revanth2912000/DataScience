```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("Online retail.xlsx",header=None)
```

```python
data.describe()
```

```python
data
```

```python
dataSplit=data[0].str.split(",",expand=True)
dataSplit
```

```python
byte = pd.get_dummies(dataSplit.stack()).groupby(level=0).sum()
byte
```

```python
for i in byte.columns:
    byte.loc[byte[i]>1]=1
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequency=apriori(byte,min_support=0.01,use_colnames=True)
frequency
```

```python
association=association_rules(frequency,metric='lift',min_threshold=2.5,num_itemsets=3)
association
```

```python
association.sort_values('lift',ascending=False)[0:30]
```

```python
association[association.lift>1]
```

# Interview Questions

What is lift and why is it important in Association rules?

Measures the strength of a rule. Lift > 1 indicates a strong association. It's important because it shows how much more often the antecedent and consequent occur together than expected by chance

What is support and Confidence. How do you calculate them?

Support: Proportion of transactions containing both items. $$ Confidence(X => Y) = (Number of transactions containing X and Y) / (Total number of transactions) $$
Confidence: Proportion of transactions with the antecedent that also have the consequent. Confidence(X => Y) = (Number of transactions containing X and Y) / (Number of transactions containing X)

What are some limitations or challenges of Association rules mining?

Limitations:
Scalability: Large datasets are hard to process.
Redundancy: Many irrelevant rules.
Interpretability: Difficult to understand many rules.
Sparsity: Hard to find meaningful associations in large itemsets.
